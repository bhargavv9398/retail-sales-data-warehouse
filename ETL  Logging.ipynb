```python
%sql
USE CATALOG `retail-sales-data-warehouse`;

CREATE SCHEMA IF NOT EXISTS logging;
```

```python
%sql
CREATE TABLE IF NOT EXISTS logging.etl_log (

    log_id BIGINT GENERATED ALWAYS AS IDENTITY,

    layer_name STRING,
    table_name STRING,

    process_start_time TIMESTAMP,
    process_end_time TIMESTAMP,

    records_processed BIGINT,

    process_status STRING,
    remarks STRING,

    created_at TIMESTAMP

)

USING DELTA;
```

```python
%sql
INSERT INTO logging.etl_log (

    layer_name,
    table_name,

    process_start_time,
    process_end_time,

    records_processed,

    process_status,
    remarks,

    created_at

)

SELECT

    'BRONZE',
    'customers_raw',

    CURRENT_TIMESTAMP(),
    CURRENT_TIMESTAMP(),

    COUNT(*),

    'SUCCESS',
    'Bronze customer data loaded successfully',

    CURRENT_TIMESTAMP()

FROM bronze.customers_raw;
```
