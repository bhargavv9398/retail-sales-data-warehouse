```python
%sql
USE CATALOG `retail-sales-data-warehouse`;
CREATE SCHEMA IF NOT EXISTS bronze;
USE SCHEMA bronze;
CREATE TABLE IF NOT EXISTS customers_raw
USING CSV
LOCATION 's3://retail-sales-data-warehouse/raw/customers_src_09052026120000.csv'
OPTIONS (header = 'true', inferSchema = 'true');
CREATE TABLE IF NOT EXISTS products_raw
USING CSV
LOCATION 's3://retail-sales-data-warehouse/raw/products_src_20042026100105.csv'
OPTIONS (header = 'true', inferSchema = 'true');
CREATE TABLE IF NOT EXISTS stores_raw
USING CSV
LOCATION 's3://retail-sales-data-warehouse/raw/stores_src_20042026100107.csv'
OPTIONS (header = 'true', inferSchema = 'true');
```

```python
%sql
CREATE TABLE IF NOT EXISTS sales_raw
USING CSV
LOCATION 's3://retail-sales-data-warehouse/raw/sales_transactions_src_20042026100107.csv'
OPTIONS (header = 'true', inferSchema = 'true');
```

```python
%sql
SELECT 'customers' AS table_name, COUNT(*) AS row_count FROM customers_raw
UNION ALL
SELECT 'products',  COUNT(*) FROM products_raw
UNION ALL
SELECT 'stores',    COUNT(*) FROM stores_raw
UNION ALL
SELECT 'sales',     COUNT(*) FROM sales_raw;
```
