```python
%sql
SELECT 'bronze_customers' AS table_name,
       COUNT(*) AS row_count
FROM `retail-sales-data-warehouse`.bronze.customers_raw

UNION ALL

SELECT 'silver_customers',
       COUNT(*)
FROM `retail-sales-data-warehouse`.silver.customers_silver

UNION ALL

SELECT 'gold_customers',
       COUNT(*)
FROM `retail-sales-data-warehouse`.gold.dim_customer;
```

```python
%sql
SELECT 'bronze_products', COUNT(*) FROM `retail-sales-data-warehouse`.bronze.products_raw

UNION ALL

SELECT 'silver_products', COUNT(*) FROM `retail-sales-data-warehouse`.silver.products_silver

UNION ALL

SELECT 'gold_products', COUNT(*) FROM `retail-sales-data-warehouse`.gold.dim_product;
```

```python
%sql
SELECT 'bronze_sales', COUNT(*) FROM `retail-sales-data-warehouse`.bronze.sales_raw

UNION ALL

SELECT 'silver_sales', COUNT(*) FROM `retail-sales-data-warehouse`.silver.sales_silver

UNION ALL

SELECT 'gold_sales', COUNT(*) FROM `retail-sales-data-warehouse`.gold.fact_sales;
```

```python
%sql
SELECT

    b.CustomerID,
    b.CustomerName,

    s.customer_id,
    s.customer_name

FROM `retail-sales-data-warehouse`.bronze.customers_raw b

JOIN `retail-sales-data-warehouse`.silver.customers_silver s

ON b.CustomerID = s.customer_id

LIMIT 10;
```

```python
%sql
DESCRIBE `retail-sales-data-warehouse`.silver.customers_silver;
```

```python
%sql
SELECT *
FROM `retail-sales-data-warehouse`.silver.customers_silver
WHERE customer_name != TRIM(customer_name);
```

```python
%sql
SELECT *
FROM `retail-sales-data-warehouse`.silver.customers_silver
WHERE email != LOWER(email);
```

```python
%sql
SELECT *
FROM `retail-sales-data-warehouse`.silver.products_silver
WHERE product_name != INITCAP(product_name);
```

```python
%sql
SELECT
    quantity,
    unit_price,
    total_amount,
    (quantity * unit_price) AS expected_total

FROM `retail-sales-data-warehouse`.gold.fact_sales

WHERE total_amount != (quantity * unit_price);
```

```python
%sql
SELECT *
FROM `retail-sales-data-warehouse`.silver.sales_silver
WHERE transaction_date IS NULL;
```

```python
%sql
SELECT
    customer_id,
    COUNT(*)

FROM `retail-sales-data-warehouse`.silver.customers_silver

GROUP BY customer_id

HAVING COUNT(*) > 1;
```

```python
%sql
SELECT *
FROM `retail-sales-data-warehouse`.gold.dim_customer
WHERE customer_name IS NULL;
```

```python
%sql
SELECT *
FROM `retail-sales-data-warehouse`.silver.sales_silver
WHERE quantity <= 0;
```

```python
%sql
SELECT *
FROM `retail-sales-data-warehouse`.gold.fact_sales f

LEFT JOIN `retail-sales-data-warehouse`.gold.dim_customer c
ON f.customer_id = c.customer_id

WHERE c.customer_id IS NULL;
```

```python
%sql
SELECT
    customer_id,
    customer_name,
    city,
    is_active,
    start_date,
    end_date

FROM `retail-sales-data-warehouse`.gold.dim_customer

ORDER BY customer_id;
```

```python
%sql
SELECT
    customer_id,
    COUNT(*)

FROM `retail-sales-data-warehouse`.gold.dim_customer

WHERE is_active = 1

GROUP BY customer_id

HAVING COUNT(*) > 1;
```

```python
%sql
SELECT *
FROM `retail-sales-data-warehouse`.gold.dim_customer
WHERE customer_id = 1
ORDER BY start_date;
```
